### Q. 2) In which year the highest number of TV Shows & Movies were released? Show with Bar Graph.

```python
df_year=pd.DataFrame(df.groupby(['type'])[['release_year']].value_counts())
```

```python
df_year
```

```python
df_year.rename(columns={0:'counts'},inplace=True)
```

```python
df_year.reset_index(inplace=True)
```

```python
df_year
```

```python
df_movie=df_year[df_year['type']=='Movie'].sort_values(by='counts',ascending=False).head(10)
```

```python
df_show=df_year[df_year['type']=='TV Show'].sort_values(by='counts',ascending=False).head(10)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
sns.barplot(x='release_year',y='counts',data=df_show,ax=axes[0])
sns.barplot(x='release_year',y='counts',data=df_movie,ax=axes[1])
axes[0].set_title('TV Show')
axes[1].set_title('Movie')
axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=90)
axes[1].set_xticklabels(axes[0].get_xticklabels(), rotation=90)
plt.tight_layout()
plt.show()
```

- From above bar graph 1 we see that in year 2020 highest number of TV Show was released. 
- In bar graph 2 we se that in year 2017 highest number of Movie was released.

### Q. 10) What is the maximum duration of a Movie/Show on Netflix ?

```python
# Convert 'duration' column to string
df['duration'] = df['duration'].astype(str)

# Extract the numerical values from 'duration' column and convert them to minutes
df['duration'] = df['duration'].str.extract('(\d+)')
df['duration'] = pd.to_numeric(df['duration'], errors='coerce')

# Convert 'Seasons' to minutes based on an assumed average episode length
average_episode_length = 30  # Assuming average episode length of 30 minutes
df.loc[df['duration'].isna(), 'duration'] = df.loc[df['duration'].isna(), 'duration'].apply(lambda x: x * average_episode_length)

# Find the maximum duration
max_duration = df['duration'].max()

print("Maximum duration of a Movie/Show on Netflix:", max_duration, "minutes")
```
